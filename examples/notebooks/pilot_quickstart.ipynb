# Orchid Ranker Pilot Quickstart
This notebook demonstrates loading a sample interactions dataframe, fitting an `OrchidRecommender`, and recording basic metrics/audit events.

```python
import pandas as pd
from orchid_ranker import OrchidRecommender, record_training, start_metrics_server, AuditLogger

# synthetic interactions (replace with customer dataset)
interactions = pd.DataFrame({
    'user_id': [1, 1, 2, 2, 3, 3],
    'item_id': [10, 11, 10, 13, 12, 14],
    'label': [1, 0, 1, 1, 0, 1],
})

# fit recommender
rec = OrchidRecommender(strategy='als', epochs=3)
rec.fit(interactions, rating_col='label')
recommendations = rec.recommend(user_id=1, top_k=3)
recommendations
```

```python
# start prometheus exporter (optional)
start_metrics_server(port=9090)
record_training(duration_seconds=2.3, epsilon=0.8)

# log DP event locally
audit_logger = AuditLogger.from_env(path='pilot_audit.jsonl')
audit_logger.log_event('pilot_setup', actor='notebook', payload={'notes': 'demo run'})
'metrics server running, audit event recorded'
```
